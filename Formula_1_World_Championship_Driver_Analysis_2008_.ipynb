# Formula 1 World Championship Driver Analysis

```python
# CSV file import
import csv
import urllib.request
import io

# List to store data and avoid rereading the CSV file every time
data_f1 = []

# URL of the CSV file on GitHub.
file_path = "https://raw.githubusercontent.com/Alessio8-ai/Project-files/refs/heads/main/formula1_data.csv"

try:
    # I open the URL and read the content
    with urllib.request.urlopen(file_path) as response:
        # Decode the contents from bytes to string
        csv_content = response.read().decode('utf-8')

    # Use io.StringIO to treat the string as a file
    formula1_file = io.StringIO(csv_content)

    # Use DictReader to read CSV into a dictionary list, making it easy to access data by column name.
    csv_reader = csv.DictReader(formula1_file)

    # Encoding the print to give it a tabular structure
    print("%s\t %s\t %s\t %s\t %s\t" % tuple(csv_reader.fieldnames))

    # Encoding the loop to give it a tabular structure
    for row in csv_reader:
        print("%s\t%s\t%s\t%s\t%s\t" % (row["Driver"], row["Team"], row["Race"], row["Country"], row["Position"]))

        # We save the row (dictionary) in our data_f1 list
        data_f1.append(row)

except Exception as e:
    print(f"Error importing CSV file from GitHub: {e}")
    print("Make sure the URL is correct and accessible, and that the file exists.")
```

```python
# Scoring system
position_points = {1: 10, 2: 8, 3: 6, 4: 5, 5: 4, 6: 3, 7: 2, 8: 1}
# Position 0 or =>9 zero points are awarded
```

```python
"""
1. Function for analyzing individual driver performances
"""

def performance_driver(driver_name):
  # I initialize the counters to ensure that the calculation starts from zero for each pilot
  total_points = 0
  victories = 0
  podiums = 0

  for race in data_f1:
    if race['Driver'].strip().lower() == driver_name.strip().lower():
            position = int(race['Position'])
            # Total points accumulated
            points = position_points.get(position, 0)
            total_points += points
            # Win count
            if position == 1:
                victories += 1
            # Podium count (1st, 2nd or 3rd place)
            if 1 <= position <= 3:
                podiums += 1

  return [total_points, victories, podiums]
```

```python
"""
Function used as key to sort by the second value (points) of each tuple (name, points)
"""

def order_key(element):
    return element[1]
```

```python
"""
2. Function for creating the final ranking of the drivers
"""

def Drivers_Standings():
    drivers_points = {}  # Dictionary (driver: points)

    # Creating the driver_score dictionary (Name: Score)
    for race in data_f1:
      driver = race["Driver"]
      position = int(race["Position"])
      points = position_points.get(position, 0)

      if driver not in drivers_points:
            drivers_points[driver] = 0
      drivers_points[driver] += points

    #  Sort the drivers by total score in descending order to generate the final standings.
    drivers_standing = list(drivers_points.items())
    drivers_standing.sort(key=order_key, reverse=True)

    # Save to file
    with open("Drivers_Standing_2008.txt", "w") as f:
      f.write("Drivers Standings 2008 Formula 1\n")
      for position, (driver, points) in enumerate(drivers_standing, start=1):
        f.write(f"{position}. {driver}: {points}\n")

    return drivers_standing, drivers_points
```

```python
"""
3. Function for creating the final constructors' ranking
"""

def Constructors_Standings(driver_points_dict, current_data):
 driver_team = {}    # Dictionary (driver: team)
 team_points = {}    # Dictionary (team: points)

  # Creating the driver_team dictionary (driver:team)
 for row in current_data:
    driver = row["Driver"]

    if driver in driver_points_dict and driver not in driver_team:
      driver_team[driver] = row["Team"]

  # Creating a team_points dictionary (team: points)
 for driver, drivers_points in driver_points_dict.items():
    if driver in driver_team:
      team = driver_team[driver]

      if team not in team_points:
        team_points[team] = 0
      team_points[team] += drivers_points

  # Sorting to create the ranking
 team_standings = list(team_points.items())
 team_standings.sort(key = order_key, reverse=True)

 return team_standings
```

```python
"""
Input and print request for Function 1
"""

# Prompt user for pilot name
driver_input = input("Enter a pilot's name (ex. Hamilton): ")
performance = performance_driver(driver_input)

# Instruction if you enter a name not present in the file
if performance[0] == 0 and performance[1] == 0:
    print(f"No data found for the pilot: {driver_input}")
else:
  # Performance Analysis Printout
    print(f"Performance for {driver_input}:")
    print(f"Total points: {performance[0]}")
    print(f"Wins: {performance[1]}")
    print(f"Podiums: {performance[2]}")
```

```python
"""
Print for Function 2
"""

# Print to inform the user that the file has been created
print("File 'Drivers_Standings_2008.txt' successfully created!\n")

# Print the final drivers' standings on the screen
final_drivers_standings = Drivers_Standings()
print("Final Drivers' Standings\n")
for position, (driver, points) in enumerate(final_drivers_standings[0], start=1):
  print(f"{position}. {driver}: {points}")
```

```python
"""
Print for Function 3
"""

# Print the final Constructors' standings on screen
final_team_standings = Constructors_Standings(final_drivers_standings[1], data_f1)
print("Final Constructors' Standings\n")
for position, (team, points) in enumerate(final_team_standings, start=1):
  print(f"{position}. {team}: {points}")
```
